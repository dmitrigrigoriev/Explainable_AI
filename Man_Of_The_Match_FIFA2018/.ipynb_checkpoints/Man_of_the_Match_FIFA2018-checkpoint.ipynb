# Model-Agnostic Method for AI

In this notebook I'll try to explain with some methods why, how the "Man of the match" prize was given and also the most important features. For the classification Random Forest will be used.

For this I'll use officcial FIFA statistic from FIFA World Cup 2018

## Import all needed libraries.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

import eli5
import eli5.sklearn
import pdpbox
import pdpbox.pdp
import shap

import matplotlib.pyplot as plt
```

```python
shap.initjs()
```

```python
data = pd.read_csv('FIFA 2018 Statistics.csv')
y = (data['Man of the Match'] == "Yes") # convert from string "Yes/No" to binary value 1/0

feature_names = [i for i in data.columns if data[i].dtype in [np.int64]] # take only interger values for our research
X = data[feature_names]
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, shuffle = True, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
```

## Measure Performance on test set

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

## Ranking features by Permutation importance

This method answers the question what are the most important features?
First we use to answer this question our test set.

Basically it answers the question "What?" (What is is most importance features?)

```python
pi_test = eli5.sklearn.PermutationImportance(model, random_state=10).fit(X_test, y_test)
eli5.show_weights(pi_test, feature_names = X_test.columns.to_list()) # Show feature importance
```

In this table features are ranked by their relative importance. Basically now we are able to answer our question "What are the most importance features?". As we can see the "Goal Scored" is the most important feature that our RandomForest-Model uses to predict the "Man of the Match" value.

I will omit the details how the feature importance is calcualted. 

But very important: **Permutation Importance will underestimate highly correlated features.**

Let's do the same with train set

```python
pi_train = eli5.sklearn.PermutationImportance(model, random_state=42).fit(X_train, y_train)
eli5.show_weights(pi_train, feature_names = X_train.columns.to_list()) # Show feature importance
```

Now the tables looks a little bit different. We have no negative values.
What does it mean? Negative values can be show as a case of bad generalization and is a form of overfitting.

General advice: **Use Permutation Importance always on test set**

## Partial Dependence Plot

Permutation Importance method is quite powerful, but we cannot still answer our another question "How?". Now we will try to answer the question "How goal scoared may change the prediction?"

We will use most important features "Goal Scored" and "Attempts"

```python
feature = 'Goal Scored'
pdp_goal_scored = pdpbox.pdp.pdp_isolate(model=model, dataset=X_test, model_features=feature_names, feature=feature)
pdpbox.pdp.pdp_plot(pdp_goal_scored, feature)
plt.show()
```

On this plot "Goal Scored" we see that if at least 1 goal was scored the the probability increates strong, but after that the line is quite flat. Basically scoring a lot of goals doesn't change to match the overall prediction of having "Man of the match" prize.

```python
feature = 'Attempts'
pdp_attempts = pdpbox.pdp.pdp_isolate(model=model, dataset=X_test, model_features=feature_names, feature=feature)
pdpbox.pdp.pdp_plot(pdp_attempts, feature)
plt.show()
```

Here we see that after 10 attempts the probability rises but than again almost no changes. Quite interesting if the team made more then 21 attempts the probability of the prize even goes down. So the answer would be to get a prize the team should make between 10 and 21 attempts.

### Interaction between features

Now we just try to answer the question how the both our most important features (on test set) change the probability of win "Man of the Match" prize.

We use "Goal Scored" and "Attempts" accordernly to our "Permutation Importance" table with **test set**.

```python
features = ['Goal Scored', 'Attempts']
pdp_interact = pdpbox.pdp.pdp_interact(model=model, dataset=X_test, model_features=feature_names, features=features)
pdpbox.pdp.pdp_interact_plot(pdp_interact_out=pdp_interact, feature_names=features)
plt.show()
```

What does it mean? Here we see: the highest probability has yellow color. It means that it should be scored about 4 goals and made about 12.5 attempts or score more then 3.5 goals and made between 16 and 18 attempts. 

If we compare with "Goal Scored" pdp plot we weill see that scoring more then 1 goals with connecting to attempts increases probability of winning "Man of the Match" prize significantly.

So intercation plots show us that we should also pay attetion for high correlated features otherwise we can our question "How?" not correct.

# SHAP (Shapley Additive exPlanations)

SHAP helps us to answer the question what in one specific match the prize "Man of the Match" was given.
To show this, we have to choose one game. I've choosen "Uruguay-Russia" from the test set.

```python
match = X_test.iloc[12]
match_array = match.values.reshape(1,-1)
model.predict_proba(match_array)
```

So we see that our model prediction of having Uruguay assigned the "Man of the Match" prize is 79%

```python
shap_explainer = shap.KernelExplainer(model.predict_proba, X_train)
shap_explainer_values = shap_explainer.shap_values(match)
shap.force_plot(shap_explainer.expected_value[1], shap_explainer_values[1], match)
```

Here we see that "Goal Scored = 3" has the most impact for the prediction, since it pushes final value to the right with biggest interval

### Summary plot

We can also see the feature impact of all the variables on the output by using SHAP summary plot

```python
shap_summary_values = shap_explainer.shap_values(X_test)
shap.summary_plot(shap_summary_values[1], X_test)
```

```python
shap.summary_plot(shap_summary_values[1], X_test, plot_type='bar')
```

This SHAP summary plots show the impact of each feature on the prediction. Goal Scored and Corners (but I would expect here Attempts) are the biggest players in this prediction. Also, getting yellow or red card literally has no effect on the result. With SHAP summary plot, it is more intuitive to tell the impact of each feature.

### SHAP Dependence Contribution Plots

```python
# If you don't supply an argument for interaction_index, Shapley uses some logic to pick one that may be interesting.
shap.dependence_plot('Attempts', shap_summary_values[1], X_test, interaction_index="Goal Scored")
```

Also, we just show that it is also capable of demonstrating the interaction of two features using SHAP values. Goal scored would have a positive influence on the SHAP values.
